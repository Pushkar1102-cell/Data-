**Box and Whisker Rating Analysis By Order Mode**

```python
sns.boxenplot(x='online_order',y='rate',data=df)
plt.xlabel('Order Mode')
plt.ylabel('rating')
plt.title('Order Mode Analysis')
plt.show()
```

**Conclusion** :If online orders ("Yes") have higher ratings:
Customers prefer establishments that offer online orders.
Businesses should focus on providing and improving online ordering services.

If offline orders ("No") have higher ratings:
Customers value in-person experiences or services over convenience.
Businesses should enhance the offline experience to attract more customers.

If both categories are similar:
Ratings might not be strongly influenced by the mode of order but by other factors like food quality, delivery time, or ambiance.

```python
pivot_table = df.pivot_table(df,index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,cmap='YlGnBu',annot=True,fmt='d')
plt.show()
```

**Conclusion** :Online Order Trends:

If "Quick Bites" or "Cafes" dominate the "Yes" column, it shows these establishments are highly reliant on online orders.
Businesses in these types should focus on enhancing their online order experience to maintain competitiveness.

Offline Preferences:
If "Fine Dining" or similar categories dominate the "No" column, it suggests that customers prefer the in-person dining experience for such types.
Strategies here could include elevating the in-restaurant experience.
Overall Pattern:

If most types lean heavily toward "Yes", it reflects a strong overall trend toward online ordering.
If "No" is dominant, certain types may still resist or not benefit from online ordering.
